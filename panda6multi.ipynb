```python
#creating multi indexes using pandas
```

```python
import numpy as np
import pandas as pd
```

```python
indexes=pd.MultiIndex.from_product([['css','cse'],[2022,2023,2024,2025]])
```

```python
##creating multiindex series
data=pd.Series([22,23,34,67,90,76,12,32],index=indexes)
```

```python
##fetching specific data in it
data[('css',2025)]
```

```python
##fetching
data['cse']
```

```python
#unstack  is used to convert multi series to dataframe #like wise stack for opposite
data.unstack()
```

```python
##creating multiindex data frame
dt=pd.DataFrame([[1,2],[3,4],[5,6],[9,7],[8,2],[9,7],[8,2],[0,2]],index=indexes)
dt.columns=['salary','total']
dt
```

```python
#similary we can create multi cols in series
##multi rows and cols as well
cols= pd.MultiIndex.from_product([['Mumbai','Delhi'],['Salary','stu']])
dts=pd.DataFrame([[1,2,8,9],[3,5,9,7],[9,7,8,9],[6,8,8,6],[3,2,6,6],[9,3,8,9],[3,2,0,5],[9,3,4,8]],columns=cols,index=indexes)
```

```python
#fetching
dts['Mumbai']["Salary"]

```

```python
#unstack converts rows to cols and cols to row by stack but it cant be applied in simple series
#dts.unstack().unstack()
dts.unstack()
```

```python
dts.stack()
```

```python
#like head tail sample
#info describe also work in data frame
dts.duplicated()
```

```python
##extracting specific rows and columns in dataframe
#dts.loc[('css',2022),('Mumbai','Salary')]
#dts.loc[:,('Mumbai','stu')]
dts
```

```python
##we need to sort index for filtering rows and cols for effieciency
dts=dts.sort_index()
dts.loc[('cse',2024):('css',2025):2]
```

```python
#filtering cols

dts.loc[:,('Mumbai','stu')]
```

```python
dts.loc[[('cse',2022),('css',2022)],[('Mumbai','Salary'),('Delhi','stu')]]
```

```python
##you can runn transpose in dataframe
dts.transpose()
```

```python
#we can swap index level as well
dts.swaplevel() #simply done on cols by pasiing axis a s well
```
