```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('/content/movies.csv')
```

```python
movies.groupby('genres')['imdb_rating'].mean().sort_values(ascending=False).head(3)
```

```python
##group by othere method
x=movies.groupby('genres')
x.first() ##it will provide 1st movie of each genre
#last
x.nth(8)
```

```python
##get group used to view all movie of group
x.get_group('Action')
```

```python
x.describe()
```

```python
x.nunique()   ##gives no of unique values
```

```python
##can use multiple aggregate function on different cols in groupby
x.agg(
    {

    'imdb_rating':['mean','sum'],
    'imdb_votes':'std'

}  )

```

```python
x.agg['mean','std'] ##applies in all cols
```

```python
#split combin eapply used to apply func in group
def find(gr):
 mask=gr['title_x'].str.startswith('The')
 return gr[mask]
```

```python
x.apply(find)
```

```python
def noramlize(gr):
  gr['normal']=(gr['imdb_rating']-gr['imdb_rating'].min())/(gr['imdb_rating'].max()-gr['imdb_rating'].min())
  return gr
```

```python
x.apply(noramlize)
```
